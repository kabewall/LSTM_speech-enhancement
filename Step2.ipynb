# Step2
音を聞いてみるステップ

ここでは作成済みのモデルと処理済みのデータセットを用いるのでGPUがなくても動作します

```python
import torch
import torch.utils.data as utils
from IPython.display import display,Audio

import model as mm
import step_utils as ut

datasets_save_dir = "./datasets"
split = [500,2000,12500] #test/val/train
batch_size = 10
sample_rate = 16000
#model parameter
num_layer = 3

model_dir_path = "./model/"

print("#####################################################################")
print("Step2 Listen Audio")
print("#####################################################################")

tensor_speech = torch.load(datasets_save_dir+"/tensor_speech")
tensor_addnoise = torch.load(datasets_save_dir+"/tensor_addnoise")

mydataset = utils.TensorDataset(tensor_speech,tensor_addnoise)
test_dataset,val_dataset,train_dataset = utils.random_split(mydataset,split)
    
train_loader = utils.DataLoader(train_dataset,batch_size=batch_size,pin_memory=True)
val_loader = utils.DataLoader(val_dataset,batch_size=batch_size,pin_memory=True)
test_loader = utils.DataLoader(test_dataset,batch_size=batch_size,pin_memory=True)
```

## いろんなモデルで聞きくらべたいときはここから

```python
model_path = 'model20201128_214536_Epoch30.pt'

# model
feat = tensor_addnoise.shape[1]
sequence = tensor_addnoise.shape[2]
model = mm.Net(sequence, feat, num_layer)
model.load_state_dict(torch.load(model_dir_path+model_path))

it = iter(test_loader)
speech,addnoise = next(it)

model.eval()

mask = model(addnoise.float())
h_hat = mask * addnoise

for i in range(h_hat.shape[0]):
    print("Index",i+1)
    addnoiseaudio = ut.tensorstft2audio(addnoise[i,:,:])
    audio = ut.tensorstft2audio(h_hat[i,:,:])
    display(Audio(addnoiseaudio,rate = sample_rate))
    display(Audio(audio,rate = sample_rate))
```
